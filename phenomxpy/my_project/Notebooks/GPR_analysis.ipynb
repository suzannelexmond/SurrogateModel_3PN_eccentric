### GPR Analysis
For each GP, the pahse and amplitude are predicted at the empirical nodes. Then for each test point I computed the true node value $y_{true}$, the mean GP prediction $\mu_{GP}$ and the GP predicted standard deviation $\sigma_{GP}$. This is not done for the reduces basis vectors, but for a uniform parameter space! These tests are done to get useful information on the harder to predict regions of parameters space and the possible shape of the distributions on the predictions (how gaussian is the error?). This can be used to calculate the standardized residuals, defined as:
$$
z = \frac{y_{true} - \mu_{GP}}{\sigma_{GP}}
$$

The perfect situation would show (perfect prediction) residuals that:
- Are centered around 0 (resembles the ground truth)
- Roughly gaussian (because the GP is supposed to be Gaussian)
- Variance around 1 
- No strong left-over structure

Explanation of the analysis figures:

## Axes
- **X-axis:** Node index (e.g., node 15, 14, ...)
- **Y-axis:** Diagnostic values:
  - |skew|
  - |excess kurtosis|
  - RMSE
  - −log10(p-value)

## Interpretation
Identifies the most problematic nodes and why:
- Skew → asymmetry in residuals
- Kurtosis → heavy tails / outliers
- RMSE → prediction accuracy
- p-value → deviation from Gaussianity

## Perfect GP
- Skew ≈ 0  
- Kurtosis ≈ 0  
- RMSE small  
- p-values large (→ −log10(p) ≈ 0)

e \rightarrow y(t_j)
\]

- Shaded region → GP ±2σ  

## Perfect GP
- GP mean lies exactly on data  
- Uncertainty bands are tight and correctly cover points
